pyGeoIGA can do refinement by knot insertion. But degree elevation depends on igakit library. Therefore is optional


```python
import numpy as np
```

<a id='knot_insertion'></a>
## 1) Knot insertion

```python
cp = np.asarray([[[0,0],[-1,0],[-2,0]],
                 [[0,1], [-1,2], [-2,2]],
                 [[1, 1.5], [1,4],[1,5]],
                 [[3, 1.5],[3, 4],[3, 5]]])
U = np.asarray([0, 0, 0, 0.5, 1, 1, 1])       
V = np.asarray([0, 0, 0, 1, 1, 1])
```

```python
from pygeoiga.nurb.nurb_creation import NURB
import matplotlib.pyplot as plt
```

```python
nrb = NURB(cp, [U, V])
nrb.knots
```

```python
fig, [ax1, ax2] = plt.subplots(1, 2, figsize=(15,5))
nrb.plot_cpoints(ax = ax1, color="red")
nrb.plot_knots(ax=ax2, color="black")
```

```python
knot_insert = [0.1, 0.2, 0.3, 0.6, 0.7]
nrb.knot_insert(knot_insert, direction = 0)
nrb.knots
```

```python
fig, [ax1, ax2] = plt.subplots(1, 2, figsize=(15,5))
nrb.plot_cpoints(ax = ax1, color="red")
nrb.plot_knots(ax=ax2, color="black")
```

```python
nrb.knot_insert(knot_insert, direction = 1)
nrb.knots
```

```python
fig, [ax1, ax2] = plt.subplots(1, 2, figsize=(15,5))
nrb.plot_cpoints(ax = ax1, color="red")
nrb.plot_knots(ax=ax2, color="black")
```

<a id='degree_elevation'></a>
## 2) Degree elevation


```python
nrb = NURB(cp, [U, V])
print(nrb.knots)
print(nrb.degree)
fig, [ax1, ax2] = plt.subplots(1, 2, figsize=(15,5))
nrb.plot_cpoints(ax = ax1, color="red")
nrb.plot_knots(ax=ax2, color="black")
```

```python
nrb.degree_elevate(times=1, direction=0)
print(nrb.knots)
print(nrb.degree)
```

```python
fig, [ax1, ax2] = plt.subplots(1, 2, figsize=(15,5))
nrb.plot_cpoints(ax = ax1, color="red")
nrb.plot_knots(ax=ax2, color="black")
```

```python
nrb.degree_elevate(times=1, direction=1)
print(nrb.knots)
print(nrb.degree)
```

```python
fig, [ax1, ax2] = plt.subplots(1, 2, figsize=(15,5))
nrb.plot_cpoints(ax = ax1, color="red")
nrb.plot_knots(ax=ax2, color="black")
```

<a id='k_refinement'></a>
## 3) K-refinement

```python
nrb = NURB(cp, [U, V])
print(nrb.knots)
print(nrb.degree)
fig, [ax1, ax2] = plt.subplots(1, 2, figsize=(15,5))
nrb.plot_cpoints(ax = ax1, color="red")
nrb.plot_knots(ax=ax2, color="black")
```

```python
knot_insert = [0.1, 0.2, 0.3, 0.6, 0.7]
nrb.knot_insert(knot_insert, direction = 0)
nrb.knot_insert(knot_insert, direction = 1)
nrb.degree_elevate(times=1, direction=0)
nrb.degree_elevate(times=1, direction=1)
print(nrb.knots)
print(nrb.degree)
```

```python
fig, [ax1, ax2] = plt.subplots(1, 2, figsize=(15,5))
nrb.plot_cpoints(ax = ax1, color="red")
nrb.plot_knots(ax=ax2, color="black")
```
